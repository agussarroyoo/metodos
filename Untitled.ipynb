```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(1,10,100)
y1 = (5-2*x)/3
y2 = (5.1-2*x)/3.1

y3 = (5.001-2*x)/3
y4 = (5.1-2*x)/3.1

plt.plot(x,y1,x,y2,x,y3,x,y4)


plt.title("Graficas")
plt.legend(('y1','y2','y3','y4'), loc='upper left')
plt.xlabel('x')
plt.ylabel('y', fontsize=20)
plt.show()
```
